```python
# subplot : 하나의 페이지에 여러개의 그래프 그림
x1 = np.linspace(0.0, 5.0) # 1부터 0.1씩증가 5까지
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2 * np.pi) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1) #(행의 수, 열의 수, 위치(1부터 시작))
plt.plot(x1,y1,'yo-')
plt.title('A table of 2 subplot')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2) 

plt.plot(x2, y2, 'r.-')
plt.xlabel('time ')
plt.ylabel('Undamped')
print(ax2)
plt.show()
```

```python
# 선 그래프
# 시도별 진출입 인구수
import pandas as pd

df = pd.read_excel('dataset/시도별 전출입 인구수.xlsx',header =0)
df.head(25)
```

```python
# 데이터 전처리 작업
# 1. Nan 처리
#    전출지의 Nan 자료를 수정 -> 이전 자료의 값으로 대체
#    Nan 자료를 수정 : method = ffill:이전자료로 대체, bfill:이후자료로 대체
#    value = 값 -> 지정하는 값으로 수정
df.fillna(method='ffill', inplace=True) 
df
```

```python
# 서울특별시에서 다른지역으로 전출한 자료만 추출
df_s = df
mask = (df_s['전출지별']=='서울특별시')&(df_s['전입지별']!='서울특별시')
df_seoul = df_s[mask].copy() # 기존의 데이터에 변형을 가한 후 새로운 변수에 
#                              넣으므로 copy를 사용해서 넣는 것이 좋다.
# True 값 만 나옴
# df_seoul = df[(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')]
# print(df_seoul.head())
df_seoul.drop('전출지별',axis=1, inplace=True)

# 3. 컬럼명 또는 인덱스 재 설정
#    '전입지별'을 '전입지'로 변경
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)

#   '전입지' 컬럼을 인덱스로 변경
df_seoul.set_index('전입지',inplace=True)
df_seoul
```

```python
df = pd.read_excel('dataset/남북한발전전력량.xlsx',convert_float = True)
df.head(10)

# 북한의 자료만 가져 옴 //수력, 화력 발전량 분석
df = df.loc[5:9] #index = 5~8
df.columns

# 전력량 (억㎾h) 컬럼 삭제
df.drop('전력량 (억㎾h)',axis = 'columns',inplace=True)
df
```

```python
# 발전전력별 컬럼을 인덱스로 지정
df.set_index('발전 전력별',inplace=True)
df
```

```python
df = df.T
```

```python
# 년도별로 증감율, '합계' 컬럼명을 '총발전량'으로 변경
df.columns = ['합계','수력','화력','원자력']
df = df.rename(columns={'합계':'총발전량'})
df
```

```python
df['총발전량 -1년'] = df['총발전량'].shift(1)
# df.drop('작년 총발전량',axis = 'columns',inplace=True)
df
# dt1.rename(columns=dt1.iloc[0]) 1행을 컬럼으로 전환
# dt3 = dt2.drop(dt2.index[0]) 1행 컬럼전환 후 1행내용 삭제
```

```python
df['증감율'] = ((df['총발전량'] / df['총발전량 -1년'])-1)*100
df
```

```python
ax1 = df[['수력','화력']].plot(kind='bar',figsize=(20,10),stacked=True)
ax2 = ax1.twinx() #x축은 공유하나 y축은 공유 안함

ax2.plot(df.index, df['증감율'],ls='--',marker='o',color='red',
        label = '전년대비 증감율(%)')
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size=20)
ax1.set_ylabel('발젼량(억 Kwh)')
ax2.set_ylabel('전년 대비 증감율 (%)')

plt.title('북한 전력 발전량')
ax1.legend(loc='best',fontsize = 20)
ax2.legend(loc='upper left',fontsize = 20)
plt.show()
```

```python
# histogram : 변수 하나

df = pd.read_csv('dataset/auto-mpg.csv', header = None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

df['mpg'].plot(kind = 'hist', bins=10, figsize = (10,5), color = 'skyblue')
plt.title('mpg histogram')
plt.xlabel('mpg')
plt.show()
```

```python
# scatter : 변수 2개
df.plot(kind='scatter',x = 'weight', y='mpg', figsize=(10,5), color='blue')
plt.title("mpg - weight scatter")
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
# scatter cylender : 변수 3개 x = 'weight', y='mpg', cyliners_size
# bubble : cylinders_size로 점의 크기변경
cylinders_size = df.cylinders / df.cylinders.max() * 300
df.plot(kind='scatter',x = 'weight', y='mpg', figsize=(10,5), color='blue',
       s = cylinders_size, alpha=0.3)
plt.title("mpg - weight scatter")
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
# scatter cylender : 변수 3개 x = 'weight', y='mpg', cyliners_size
# cylinders_size를 컬러맵으로 사용
cylinders_size = df.cylinders / df.cylinders.max() * 300
df.plot(kind='scatter',x = 'weight', y='mpg', figsize=(10,5), 
       c = cylinders_size, s = 50, alpha=0.3, cmap='viridis')
plt.title("mpg - weight scatter")
plt.xlabel('weight')
plt.ylabel('mpg')
plt.savefig('img/scatter.png') # file에 그래프 저장 (img디렉토리/scatter.png)
plt.show()
```

```python
# pie chart
# 데이터 개수 카운트를 위해 값 1을 가진 열을 추가
df['count'] = 1
df_origin = df.groupby('origin').sum()# origin 열을 기준으로 그룹화, 합계 연산
print(df_origin.head())                  # 그룹 연산 결과 출력

# 제조국가(origin) 값을 실제 지역명으로 변경
df_origin.index = ['USA', 'EU', 'JAPAN']

# 제조국가(origin) 열에 대한 파이 차트 그리기 – count 열 데이터 사용
df_origin['count'].plot(kind='pie', 
                     figsize=(7, 5),
                     autopct='%1.1f%%',   # 퍼센트 % 표시
                     startangle=10, # 파이 조각을 나누는 시작점(각도 표시)
                     colors=['chocolate', 'bisque', 'cadetblue']# 색상 리스트
                     )

plt.title('Model Origin', size=20)
plt.axis('equal')    # 파이 차트의 비율을 같게 (원에 가깝게) 조정
plt.legend(labels=df_origin.index, loc='upper right')   # 범례 표시
plt.show()
```

```python
# boxplot  이상 데이터 감지(제조업체에서 제품 정상확률 구하기 위해 사용)
# 그래프 객체 생성 (figure에 2개의 서브 플롯을 생성)
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# axe 객체에 boxplot 메서드로 그래프 출력
ax1.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']], 
         labels=['USA', 'EU', 'JAPAN'])

ax2.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']], 
         labels=['USA', 'EU', 'JAPAN'],
         vert=False)# 수평으로 그리기

ax1.set_title('제조국가별 연비 분포(수직 박스 플롯)')
ax2.set_title('제조국가별 연비 분포(수평 박스 플롯)')

plt.show()
```
