```python
titanic = sns.load_dataset('titanic')

# age와 fare 컬럼만 추출
df = titanic.loc[ : ,['age','fare']]
df.head()
```

```python
# 숫자에 10을 더한 값을 구하는 함수
def add_10(n):
    return n + 10

# 두 객체의 합을 구하는 함수
def add_two_obj(a, b):
    return a + b

print(add_10(10), add_two_obj(10, 20))
```

```python
# age 컬럼의 값에 10을 더한 값을 실행
# series 에 함수 매핑 -> 시리즈.apply(함수)
df['age=10'] = df['age'].apply(add_10)

# age와 b의 값을 더함
df['age+fare'] = df['age'].apply(add_two_obj, b=20)
df.head()

df['age'].apply(lambda x: x + 10)# 간단함 함수는 lambda로 지정
```

```python
# 데이터프레임 객체에 함수 매핑
# 데이터프레임.pipe(함수)
def missing_value(x): # x -> 1개의 원소
    return x.isnull()

def missing_count(x): # x -> data frame
    return missing_value(x).sum()

def total_number_missing(x): # x-> series
    return missing_count(x).sum()
```

```python
result_value = df.pipe(missing_value)
print(type(result_value))  # 데이터프레임

result_value = df.pipe(missing_count)
print(type(result_value))  # 시리즈

result_value = df.pipe(total_number_missing)
print(type(result_value))  # 값
```

```python
# 데이터프레임 합치기 : pd.concat(데이터프레임 리스트)
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
result1 = pd.concat([df1,df2])#컬럼이 없으면 Nan
print(result1)
print()
# 기존의 인덱스를 무시하고 인덱스 재설정
result2 = pd.concat([df1,df2],ignore_index=True)
print(result2)
print()

# 열 단위로 합성
result3 = pd.concat([df1,df2], axis =1) # 인덱스가 없으면 Nan
print(result3)
print()

# 인덱스가 같은 행만 조인
result4 = pd.concat([df1,df2], axis =1, join='inner') # join='inner'인덱스가
########################################################같은것만 조인
print(result4)
print()
```
