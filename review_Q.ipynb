```python
df = pd.read_csv('dataset/auto-mpg.csv', header = None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
df.plot(kind='scatter',x = 'weight', y='mpg', figsize=(10,5), color='blue')
plt.title("mpg - weight scatter")
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
cylinders_size = df.cylinders / df.cylinders.max() * 300
df.plot(kind='scatter',x = 'weight', y='mpg', figsize=(10,5), color='blue',
       s = cylinders_size, alpha=0.3)
plt.title("mpg - weight scatter")
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
cylinders_size = df.cylinders / df.cylinders.max() * 300
df.plot(kind='scatter',x = 'weight', y='mpg',marker='+',
        figsize=(10,5), cmap ='viridis',c = cylinders_size,s=50, alpha=0.3)
plt.title("Scatter Plot: mpg-weight-cylinders")
plt.savefig('./dataset/scatter1.png')
plt.savefig('./dataset/scatter1_transparent.png', transparent=True)
plt.show()
```
