```python
# pandas에서 제공하는 그래프 함수 plot()
df = pd.read_excel('dataset/남북한발전전력량.xlsx')
df.head(7)
# 남한과 북한의 1991년 이후 자료만 가져 옴
df_ns = df.iloc[[0,5],3:]
df_ns # 0와 5인덱스명을 '0':'south','5':'north' 인덱스명을 변경
df_ns.index = ['South','North']
df_ns
# 데이터의 정보 확인
df_ns.info() 
df_ns.columns = df_ns.columns.map(int) # columns의 이름을 정수형으로 변경
df_ns.columns
```

```python
df_ns.plot() # index -> x 축, 년도를 기준으로 남한과 북한의 전력량을 확인
tdf_ns = df_ns.T #인덱스와 컬럼의 위치 변환
tdf_ns.head()
tdf_ns.plot() # 남한과 북한의 연도별 전력량을 그래프로 표시
```

```python
tdf_ns.plot(kind = 'bar')# 막대그래프
tdf_ns.plot(kind = 'hist')
```

```python
arr = np.arange(15) #->list 아님
type(arr) # 백터연산 가능
print(arr,type(arr))
arr = arr.reshape(3,5) # reshape : 배열 재정의
print(arr)
print(arr.ndim) # ndim 몇차원인지?
print(arr.shape, arr.size, arr.dtype.name) #.size -> 배열크기(요소 수)
print(arr.itemsize) # 데이터의 저장 공간 사이즈 : 4byte
```

```python
arr = np.arange(-5, 5, 0.5)
print(arr.size)
arr = arr.reshape(4,int(arr.size/4))
print(arr)
arr1 = arr.flatten() # 다차원 -> 1차원
arr1.ndim
```

```python
# 기존의 파이썬 리스트로 array 생성
list_a = np.array([1,2,3])
list_b = np.array([[1,2,3],[4,5,6]])
print('list_a :{}, list_a의 type : {} '.format(list_a,type(list_a)))
print('list_b :{}, list_b의 type : {} '.format(list_b,type(list_b)))
print('list_a dim : {}, list_a의 shape:{}'.format(list_a.ndim,list_a.shape))
print('list_b dim : {}, list_b의 shape:{}'.format(list_b.ndim,list_b.shape))
print('list_a의 size {}, list_a의 dtype {}'.format(list_a.size,list_a.dtype))
```
